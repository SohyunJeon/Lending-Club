# 3. Model Selection

### (2) Dataframe sampling and shuffle

```python
def shuffle_data(X_0, X_1): # select same number of sample with X_1 from X_0, and shuffle
    
    X_sample = X_0.sample(n=45936)
    
    X = pd.concat([X_sample, X_1])
    
    X = X.sample(frac=1).reset_index(drop=True)
    
    y = X['loan_status']
    X = X.drop('loan_status', axis=1)
    
    return X, y
```

### (3) Cross Validation

```python
def cross(model, X_0, X_1, thres):
    
    X, y = shuffle_data(X_0, X_1)
    
    train_list = []
    test_list = []
    f1_list =[]
    auc_list=[]
    
     # 5 times modeling and take score
    for i in range(5):
        X_train, X_val, y_train, y_val = train_test_split(X, y)
        
        # Training
        fitted_model = model.fit(X_train, y_train)
        
        # accuracy
        train_list.append(model.score(X_train, y_train))
        test_list.append(model.score(X_val, y_val))
        
        # AUC $ ROC
        auc_list.append(roc_auc_score(y_val, model.predict(X_val)))
        #f1
        y_pred = model.predict_proba(X_val)
        y_true = y_val
        pred_random=[]

        THRESHOLD = thres
        for val in y_pred:
            if val[1] > THRESHOLD:
                pred_random.append(1)
            elif val[1] <= THRESHOLD:
                pred_random.append(0)
        f1_list.append(f1_score(y_true, pred_random))

   
    print('train score : {} & mean : {}'.format(train_list, np.array(train_list).mean()))
    print('test score: {}  & mean : {} '.format(test_list, np.array(test_list).mean()))
    print('f1 score : {} & mean : {}'.format(f1_list, np.array(f1_list).mean()))
    print('auc&roc score : {} & mean : {}'.format(auc_list, np.array(auc_list).mean()))
    
    
    return fitted_model, X_train 
```

### Test Score

```python
def test (model, X, y, thres):
    
    # accuracy 
    test_accuracy = model.score(X, y)
     
    #f1
    y_pred = model.predict_proba(X)
    y_true = y
    pred_random=[]

    THRESHOLD = thres
    for val in y_pred:
        if val[1] > THRESHOLD:
            pred_random.append(1)
        elif val[1] <= THRESHOLD:
            pred_random.append(0)
    test_f1 = f1_score(y_true, pred_random)

     
    
    print('test score: {}'.format(test_accuracy))
    print('auc&roc score : {} '.format(roc_auc_score(y, model.predict(X))))
    print('f1 score : {} '.format(test_f1))
  
```
